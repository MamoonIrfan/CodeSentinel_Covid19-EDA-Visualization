Step 1: Correlation matrix heatmap
--------------------------

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset load karo
df = pd.read_csv("covid19_countrywise_data.csv")

# Sirf numerical columns nikal lo
numeric_df = df.select_dtypes(include=['number'])

# Correlation matrix banao
corr = numeric_df.corr()

# Heatmap plot karo
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix Heatmap (COVID-19 Data)", fontsize=14)
plt.show()
```

Interpretation of Data Visualization:
-------------------------------------
-There is a strong positive correlation between Confirmed cases and Deaths ‚Üí higher confirmed cases lead to more reported deaths.

-Confirmed cases also show a high correlation with Recovered and Active cases ‚Üí meaning recoveries and active patients increase as total cases rise.

-A direct relationship exists between New cases and New deaths ‚Üí when new cases increase, new deaths also tend to rise.

-Ratios (Deaths/100 Cases, Recovered/100 Cases) show weak or negative correlation ‚Üí these percentages do not follow the same trend as absolute numbers.

-The heatmap highlights that absolute numbers (Confirmed, Deaths, Recovered) move closely together, while ratios behave independently.

Step 3: Distribution plots (histograms, boxplots)
-----------------------------------------------------

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("covid19_countrywise_data.csv")

# 1) Top 5 most frequent values in selected columns
# Choose categorical columns from dataset
selected_columns = ["Country/Region", "WHO Region"]

for col in selected_columns:
    print(f"\nTop 5 most frequent values in column: {col}")
    print(df[col].value_counts().head(5))

# 2) Distribution plots (Histograms and Boxplots)
# Choose numerical columns for distribution analysis
numeric_columns = ["Confirmed", "Deaths", "Recovered", "Active"]

for col in numeric_columns:
    plt.figure(figsize=(12, 5))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Histogram of {col}")

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")

    plt.tight_layout()
    plt.show()
```

Interpretation of Data Visualization 
------------------------------------

1. Confirmed Cases

-Histogram shows that most countries have low confirmed cases, while very few have extremely high cases.

-Boxplot highlights the presence of strong outliers (countries with very high confirmed cases compared to the majority).

2. Deaths

-Histogram shows that deaths are highly concentrated near the lower range; only a few countries report very high deaths.

-Boxplot confirms extreme outliers in death counts (a few countries driving the maximum).

3. Recovered Cases

-Histogram indicates recovery numbers are also skewed, with most countries reporting relatively small recoveries.

-Boxplot highlights that a handful of countries have unusually high recovery counts compared to the rest.

4. Active Cases

-Histogram suggests that active cases are unevenly distributed, with most countries having small active cases and a few with very high active cases.

-Boxplot shows extreme outliers in active cases as well, proving a large imbalance between countries.

5. WHO Region (Categorical)

-Value counts show which WHO region has the highest representation of countries in the dataset.

-This helps us understand global distribution by region and focus on the most affected areas.

6. Country/Region (Categorical)

-Top 5 countries show which nations appear most frequently in the dataset.

-This can indicate either data reporting style (multiple entries per country) or special focus regions in the dataset.
